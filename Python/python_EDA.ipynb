```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.float.format', lambda x:'%.2f' % x)  # change float format
```

```python
df = pd.read_csv('data/world_population2.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()  # check for missing values
```

```python
df.nunique()  # number of unique values
```

```python
df.sort_values(by="2022 Population").head()  # sort ascending - lowest to highest
```

```python
df.sort_values(by="2022 Population", ascending=False).head(10)  # sort descending - highest to lowest
```
