# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

### Line Chart

```python
df.plot()  # line plot by default
```

```python
df.plot(kind='line', subplots=True)
```

```python
df.plot(kind='line', title="Ice Cream Ratings")
```

```python
df.plot(kind='line', title="Ice Cream Ratings", xlabel='Daily Ratings', ylabel='Scores')
```

### Bar Chart

```python
df.plot(kind='bar', title="Ice Cream Ratings")
```

```python
df.plot(kind='bar', stacked=True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot.barh(stacked=True)
```

### Scatter Plot

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500)
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=500, c='purple')
```

### Histograms

```python
df.plot.hist()
```

```python
df.plot.hist(bins=20)  # bins represents the number of tick marks in the x label.
```

### Box Plot

```python
df.boxplot()
```

### Area Chart

```python
df.plot.area(figsize=(10, 5))
```

### Pie Chart

```python
df.plot.pie(y='Flavor Rating', figsize=(10, 10))
```
